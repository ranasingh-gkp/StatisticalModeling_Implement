## 中心極限定理
平均$\mu$、分散$\sigma^2$の任意の確率分布に従う母集団から抽出した標本において、サンプルサイズが大きくなるにつれて標本平均の分布は平均$\mu$、分散$\sigma^2/n$の正規分布に近づく。

```python
%matplotlib inline

import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
from tqdm import tqdm_notebook
from scipy.stats import norm, poisson, gamma, uniform

sns.set_style('white')
sns.set_context('talk')

np.random.seed(123)
```

```python
from scipy import stats
```

```python
mu = 3.0
n_sample = 100

def poisson_means(n_sample):
    means = []
    for _ in range(1000):
        data = poisson.rvs(mu=mu, size=n_sample)
        means.append(data.mean())

    means = np.array(means)
    plt.figure()
    plt.hist(means)
```

```python
for i in [2, 10, 100, 1000]:
    poisson_means(i)
```

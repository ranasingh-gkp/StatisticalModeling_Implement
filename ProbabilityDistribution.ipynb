```python
import numpy as np
import numpy.random as rnd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### t distribution

### geometric distribution
確率pで成功する試行が初めて成功するまでの試行回数nが従う確率分布。

```python
s = rnd.geometric(p=0.2, size=10000)
sns.distplot(s, kde=False, bins=np.arange(max(s)+1))
```

### exponential distribution
ある期間に平均してラムダ回起こる現象が次に起きるまでの期間Xが従う確率分布。幾何分布の連続変数バージョン。
  
例） 機械が故障してから次に故障するまでの期間　　
  
$$f\left( x\right) =\dfrac {1}{\mu }\exp \left( -\dfrac {x}{\mu }\right)　$$
$\mu$: scale parameter

```python
s = rnd.exponential(scale=5.0, size=10000)
sns.distplot(s, kde=False, bins=30)
```

### poisson distribution
一定期間においてある事象が発生する回数が従う確率分布

* probability of a given number of events occurring in a fixed interval of time  

if these events occur
* with a known constant rate
* independently of the time since the last event  



### negative binomial distribution

成功確率pの試行がk回成功するまでに必要な試行回数xが従う確率分布。  
  
numpyの場合、成功回数 n、成功確率pとした場合の失敗回数が従う確率分布となる。  
この場合平均は$n(1-p)/p$、分散$n(1-p)/p^2$

$$p\left( x\right) =\begin{pmatrix} x-1 \\ k-1 \end{pmatrix} {p^{k}\left( 1-p\right) ^{x-k}}$$

```python
s = rnd.negative_binomial(n=1.0, p=0.2, size=1000)
sns.distplot(s, kde=False, bins=np.arange(max(s)+1))
```

### Weibull distributioin
事象の生起確率が対象とする期間内において変化する場合，その事象が発生するまでの時間を確率変数とみなすと，その確率変数が従う分布は指数分布ではなくワイブル分布となる。  
  
例）　電化製品が故障するまでの時間（経年劣化により故障率が上昇）
  
shape parameter (b)が１より大きい場合は時間経過とともに発生率が上昇、1より小さい場合は発生率が低下するモデル。  
b=1の場合は指数分布となる。

### Vonmises distribution

```python
from scipy import stats
fig, ax = plt.subplots(1, 1)    
kappa = 3.99390425811
x = np.linspace(stats.vonmises.ppf(0.01, kappa),
              stats.vonmises.ppf(0.99, kappa), 100)
ax.plot(x, stats.vonmises.pdf(x, kappa),
         'r-', lw=5, alpha=0.6, label='vonmises pdf')

rv = stats.vonmises(kappa)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
```

```python
x = np.linspace(-np.pi, np.pi, 100)
fig, ax = plt.subplots(1, 1)


def plot_vonmises(x, kappa, ax):
    ax.plot(x, stats.vonmises.pdf(x, kappa),
             '-', lw=3, alpha=0.6, label=f'kappa={kappa}')

kappas = [0.5, 1, 2]
for kappa in kappas:
    plot_vonmises(x, kappa, ax)
ax.legend()
```

```python
x = np.linspace(-np.pi, np.pi, 100)
fig, ax = plt.subplots(1, 1)

kappa = 1

def plot_vonmises(x, kappa, scale, ax):
    ax.plot(x, stats.vonmises.pdf(x, kappa, scale=scale),
             '-', lw=3, alpha=0.6, label=f'scale={scale}')

scales = [0.5, 1, 2]
for scale in scales:
    plot_vonmises(x, kappa, scale, ax)
ax.legend()
```

```python
x = np.linspace(-np.pi, np.pi, 100)
fig, ax = plt.subplots(1, 1)

def plot_vonmises(x, kappa, ax):
    ax.plot(x, stats.vonmises.pdf(x, kappa, loc=1),
             '-', lw=3, alpha=0.6, label=f'kappa={kappa}')

kappas = [0.5, 1, 2]
for kappa in kappas:
    plot_vonmises(x, kappa, ax)
ax.legend()
```

```python
x = np.linspace(-2, -1)
mu = np.arctan(-20)
kappa = 50

plt.plot(x, stats.vonmises.pdf(x, kappa, loc=mu),
         '-', lw=3, alpha=0.6, label=f'kappa={kappa}')
```
